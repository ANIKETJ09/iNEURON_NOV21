# Affine Transformations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_vec(vectors, colors):
  plt.figure(dpi=100)
  plt.axhline(y=0, color = 'lightgrey')
  plt.axvline(x=0, color='lightgrey')
  plt.xlim(-5,5)
  plt.ylim(-5,5)

  for i in range(len(vectors)):
    x = np.concatenate([[0,0],vectors[i]])
    plt.quiver(x[0],x[1],x[2],x[3],scale=1, scale_units='xy', angles='xy', color=colors[i])
```

```python
v= np.array([2,3])
```

```python
v
```

```python
plot_vec([v],['red'])
```

```python
I = np.array([[1, 0], [0, 1]])
I
```

```python
K = np.array([[1, 0], [0, -1]])
K
# It work as functional Matrix.
```

```python
vK = np.dot(v,K)
vK
```

```python
plot_vec([v,vK],['blue', 'lightblue'])
```

```python
G = np.array([[-1, 0], [0, 1]])
G 
# It work as functional Matrix.
```

```python
vG = np.dot(v, G)
vG
```

```python
plot_vec([v,vG], ['blue', 'lightblue'])
```

```python
# SHEAR TRANFORMATION

A = np.array([[-1, 2], [2, -2]])
A
# It work as functional Matrix.
```

```python
Av = np.dot(A,v)
```

```python
plot_vec([v, Av], ['green', 'lightgreen'])
```

## Singular Vector Decomposition

$ A = U D V^{T}$

* $U=$ Orthogonal $ m × m $ matrix
* $D=$ Diagonal matix of order $ m × n$
* $V=$ Orthogonal $n × n$ matix

## Image compression using SVD (Application of SVD)

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = Image.open('photo.jpg')
_ = plt.imshow(img)
```

"1" 1 bit Black and white

"L" 8 bit black and white

"LA" Alpha mode

```python
im_gray = img.convert('LA')
_ = plt.imshow(im_gray)
```

```python
im_mat = np.array(list(im_gray.getdata(band=0)), float)
im_mat
```

```python
im_mat.shape
```

```python
im_mat.shape = (im_gray.size[1], im_gray.size[0])
im_mat.shape
```

```python
im_mat = np.matrix(im_mat)
_ = plt.imshow(im_mat, cmap='gray')
```

```python
U, sigma, V = np.linalg.svd(im_mat)
```

```python
new_img = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
_ = plt.imshow(new_img, cmap='gray')
```

```python
for i in [2, 4, 8, 16, 32, 64]:
    new_img = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(new_img, cmap='gray')
    plt.title(i)
    plt.show()
```

## Application of Pseudoinverse

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [0.,1,2,3,4,5,6,7,8]
y = [12,10,15,14,16,18,19,20,22]
```

```python
figure, axes = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
_ = axes.scatter(x,y)
```

```python
x0 = np.ones(9)
x0
```

```python
X= np.concatenate((np.matrix(x0).T, np.matrix(x).T), axis=1)
X
```

```python
W = np.dot(np.linalg.pinv(X),y)
W
```

```python
c = W[0,0]
c
```

```python
m = W[0,1]
m
```

```python
from matplotlib.image import AxesImage
figure, axes = plt.subplots()

plt.xlabel('x')
plt.ylabel('y')

axes.scatter(x, y)

x_min, x_max = axes.get_xlim()
y_at_xmin = m*x_min + c
y_at_xmax = m*x_max + c

axes.set_xlim([x_min, x_max])
_ = axes.plot([x_min, x_max], [y_at_xmin, y_at_xmax], c='red')
```
