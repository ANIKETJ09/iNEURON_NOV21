```python
import torch
import matplotlib.pyplot as plt
import numpy as np
import math
```

$$ z = x^2 - y^2 $$

```python
def func(x_, y_):
  return x_**2 - y_**2
```

```python
x = np.linspace(-5,5,10000)
```

```python
z_x = func(x, 0)
```

```python
figure, axes = plt.subplots(dpi=100)
plt.axvline(x=0, color='lightgrey')
plt.axhline(y=0, color='lightgrey')

axes.plot(x,z_x)
```

$$ \frac{\partial z}{\partial x} = 2x $$

```python
z1_x = -6 *x + -9
```

```python
z2_x = 6 * x + -9
```

```python
figure, axes = plt.subplots(dpi=100)
plt.axvline(x=0, color='lightgrey')
plt.axhline(y=0, color='lightgrey')

axes.set_ylim(-2,30)

plt.scatter(-3,9, color = 'orange')
plt.scatter(3,9, color='red')
axes.plot(x, z2_x, color='red', linestyle=':')
axes.plot(x,z1_x, color = 'orange', linestyle=':')
axes.plot(x,z_x)
```

```python
y = np.linspace(-5,5,10000)
```

```python
z_y = func(0, y)
```

```python
figure, axes= plt.subplots(dpi=100)
axes.axhline(y=0, color='lightgrey')
axes.axvline(x=0, color= 'lightgrey')

_ = axes.plot(y, z_y)
```

$$ \frac{\partial z}{\partial y} = -2y $$

```python
z1_y = 6 * y + 9
```

```python
z2_y = -6 *x + 9
```

```python
figure, axes= plt.subplots(dpi=100)
axes.axhline(y=0, color='lightgrey')
axes.axvline(x=0, color= 'lightgrey')

axes.set_ylim(-30, 2)

plt.scatter(-3,-9, color = 'orange')
plt.scatter(3,-9, color='red')
axes.plot(y, z2_y, color='red', linestyle=':')
axes.plot(y,z1_y, color = 'orange', linestyle=':')

_ = axes.plot(y, z_y)
```

## Partial derivatives of scalars using PyTorch

```python
import torch
```

```python
x = torch.tensor(2., requires_grad=True)
x
```

```python
y = torch.tensor(3.).requires_grad_()
y
```

```python
z = x**2 - y**2
```

```python
z.backward()
```

```python
x.grad
```

```python
y.grad
```

## Partial derivatives of tensors using Pytorch

```python
x = torch.tensor([2,3,6.]).requires_grad_()
x
```

```python
y = torch.tensor([6,5,1.]).requires_grad_()
y
```

```python
z = x**2 - y**2
```

```python
z.sum().backward()
```

```python
x.grad
```

```python
y.grad
```

## Partial Derivatives using TensorFlow

```python
import tensorflow as tf
```

```python
x = tf.Variable(3.)
x
```

```python
y = tf.Variable(2.)
y
```

```python
with tf.GradientTape() as t:
  t.watch(x)
  z = x**2 - y**2
t.gradient(z, x)
```

```python
with tf.GradientTape() as t:
  t.watch(y)
  z = x**2 - y**2
t.gradient(z,y)
```

# Volume of Cuboid

```python
import torch
```

```python
def cuboid_vol(l_,b_,h_):
  return l_*b_*h_
```

```python
l = torch.tensor(5.).requires_grad_()
l
```

```python
b = torch.tensor(7.).requires_grad_()
b
```

```python
h = torch.tensor(9.).requires_grad_()
h
```

```python
v = cuboid_vol(l,b,h)
v
```

```python
v.backward()
```

$ \frac{ v}{ l} = bh $

```python
l.grad
```

```python
cuboid_vol(3,b,h) - cuboid_vol(2,b,h)
```

```python
cuboid_vol(6,b,h) - cuboid_vol(5,b,h)
```

```python
cuboid_vol(10,b,h) - cuboid_vol(9,b,h)
```

$ \frac{ v}{ b} = lh $

```python
b.grad
```

$ \frac{ v}{ h} = lb $

```python
h.grad
```

# Volume of Cone

```python
import math
```

```python
def cone_vol(r_,h_):
  return 1/3 * math.pi * r_**2 * h_
```

```python
r = torch.tensor(5.).requires_grad_()
r
```

```python
h = torch.tensor(8.).requires_grad_()
h
```

```python
v = cone_vol(r,h)
v
```

```python
v.backward()
```

```python
h.grad
```

```python
cone_vol(r,5) - cone_vol(r, 4)
```

```python
r.grad
```

```python
cone_vol(5,h) - cone_vol(4,h)
```

```python
delta = [1, .1, .01, .001, .0001]
```

```python
for i in delta:
  print((cone_vol(r+i, h) - cone_vol(r, h))/i)
```
