```python
import mysql.connector as conn
```

```python
mydb = conn.connect(host = "localhost" , user = "root" , passwd = "Aniket1996@")
```

```python
cursor = mydb.cursor()
```

```python
cursor.execute("show databases ")
cursor.fetchall()
```

```python
# DQL 

# 'SELECT' operation.

cursor.execute("select  * from ineuron.glassdata ")
cursor.fetchall()
```

```python
# 'select' operation.

cursor.execute("select col5, col5 , col10 , col1 from ineuron.glassdata")
cursor.fetchall()
```

```python
# ' SELECT ' operation with ' WHERE ' condition.

cursor.execute("select * from ineuron.glassdata WHERE col1 = 4 ")
cursor.fetchall()
```

```python
# 'Delete' Operation.

cursor.execute("delete from ineuron.glassdata where col1 = 4")
mydb.commit()

cursor.execute("select * from ineuron.glassdata ")
cursor.fetchall()
```

```python
# ' DELETE OPERATION '.

cursor.execute("delete from ineuron.glassdata where col1 = 1")
mydb.commit()

cursor.execute("select * from ineuron.glassdata ")
cursor.fetchall()
```

```python
# " INSERT OPERATION "
cursor.execute('insert into ineuron.glassdata values(1,1.52101,13.64,4.49,1.10,71.78,0.06,8.75,0.00,0.00,1)')
```

```python
mydb.commit()

cursor.execute("select * from ineuron.glassdata ")
cursor.fetchall()
```

```python
# 'Update Operation'.

cursor.execute(" UPDATE ineuron.glassdata SET col1 = 6 , col2 = 8  where col1  = 1")
mydb.commit()

cursor.execute("select * from ineuron.glassdata ")
cursor.fetchall()
```

```python
# DDL OPERATION [CREATE, ALTER, DELETE] ON COLUMN LEVEL.

# 1. CREATE COLUMN i.e. [ADD COLUMN] IN TABLE. FOR THIS WE REQUIRE ----|
#                                                                      |
#                                                                      |
# 2. ALTER COMMAND  <----------------------<--------------------------<-



# 2. ALTER COMMAND  <----------------------<--------------------------<-
#                                                                       |
#                                                                       |
# 3. DROP COLUMN i.e. [REMOVE COLUMN] IN TABLE.FOR THIS WE REQUIRE -----|

# ALTER COMMAND IS USED FOR:->

# 1. ADD COLUMN IN TABLE.
# 2. CHANGE DATA TYPYE OF COLUMN.
# 3. CHANGE COLUMN NAME.
# 4. CHANGING COLUMN POSTION.

# 5. REMAINING TABLES

# " Alter Operation " for column (' ADD column ') 


cursor.execute(" Alter table ineuron.glassdata ADD column coll12 varchar(20)")
mydb.commit()

cursor.execute("select * from ineuron.glassdata ")
cursor.fetchall()
```

```python
# " Drop Operation " for column ---> (' drop column ')

# step 1
# 1st  will do " Alter Operation " i.e. ---> (' ADD column ')

cursor.execute(" Alter table ineuron.glassdata ADD column coll13 varchar(20)")
mydb.commit()

cursor.execute("select * from ineuron.glassdata ")
cursor.fetchall()
```

```python
# step 2
# Then i will do " Drop Operation " i.e. ---> (' drop column ')

cursor.execute(" Alter table ineuron.glassdata drop column coll13 ") # Here after the column name no need to 
                                                                     # write its datatype.
mydb.commit()

cursor.execute("select * from ineuron.glassdata ")
cursor.fetchall()
```
